```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("cardekho_dataset.csv")
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.drop({'Unnamed: 0','car_name','brand','max_power'},axis=1,inplace=True)
```

```python
data
```

```python
data['model'].unique()
```

```python
data['transmission_type']=data['transmission_type'].replace('Manual',1)
data['transmission_type']=data['transmission_type'].replace('Automatic',2)
```

```python
data['fuel_type']=data['fuel_type'].replace('Petrol',1)
data['fuel_type']=data['fuel_type'].replace('Diesel',2)
data['fuel_type']=data['fuel_type'].replace('CNG',3)
data['fuel_type']=data['fuel_type'].replace('LPG',4)
data['fuel_type']=data['fuel_type'].replace('Electric',5)
```

```python
data['seller_type']=data['seller_type'].replace('Individual',1)
data['seller_type']=data['seller_type'].replace('Dealer',2)
data['seller_type']=data['seller_type'].replace('Trustmark Dealer',3)
```

```python
j=1
for i in data['model'].unique():    
    data['model']=data['model'].replace(i,j)
    print(i,j)
    j=j+1
```

```python
x=data[['model','vehicle_age','km_driven','seller_type','fuel_type','transmission_type','mileage','engine','seats']]
y=data['selling_price']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_train)
```

```python
model.predict([[19,1,4500,2,1,2,18.40,1498,5]])
```
